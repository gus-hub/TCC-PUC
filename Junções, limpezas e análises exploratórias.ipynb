# Transformações

## Análise de nulos e remoções

```python
nulos = pd.DataFrame(data=df_lote_devolvido_PF.isnull().sum())
nulos = nulos.loc[nulos[0]>0]
nulos['%'] = nulos[0]*100/10065
nulos.columns = ['Contagem', '% de nulos']
nulos
```

### Colunas usadas como chaves dos datasets, exceto NI Beneficiário

```python
colunas_chaves = ['Processo', 'Nº Processo', 'CPF']

df_lote_devolvido_PF = df_lote_devolvido_PF.drop(columns=colunas_chaves)
print(f'Dimensões do Lote:\n{df_lote_devolvido_PF.shape} \n')
print(f'Campos do Lote:\n{list(df_lote_devolvido_PF.columns)}\n')
```

### Coluna 'Ano' e 'Tipo_NI_Int'

```python
colunas_ano = ['Ano', 'Tipo_NI_Int']
df_lote_devolvido_PF = df_lote_devolvido_PF.drop(columns=colunas_ano)

print(f'Dimensões do Lote:\n{df_lote_devolvido_PF.shape} \n')
print(f'Campos do Lote:\n{list(df_lote_devolvido_PF.columns)}\n')
```

### Atributos constantes

```python
atributo = []
valor = []
for coluna in df_lote_devolvido_PF.columns:
    if df_lote_devolvido_PF[coluna].nunique() > 1:
        pass
    else:
        atributo.append(coluna)
        unico = df_lote_devolvido_PF[coluna].unique()[0]
        valor.append(unico)
        df_lote_devolvido_PF = df_lote_devolvido_PF.drop(columns=coluna)
        
print(f'Dimensões do Lote: {df_lote_devolvido_PF.shape}')
pd.DataFrame({'Atributo':atributo, 'Valor único':valor})
```

## Criação da coluna "Idade"

```python
df_lote_devolvido_PF['Data de Nascimento'] = pd.to_datetime(df_lote_devolvido_PF['Data de Nascimento'])
df_lote_devolvido_PF['Idade'] = 2020 - df_lote_devolvido_PF['Data de Nascimento'].dt.year
df_lote_devolvido_PF = df_lote_devolvido_PF.drop(columns='Data de Nascimento')
```

## Ano de óbito

```python
df_lote_devolvido_PF['Ano de Óbito'].unique()
```

```python
df_lote_devolvido_PF['Ano de Óbito'] = df_lote_devolvido_PF['Ano de Óbito'].str.replace('<N/D>', '0', regex=False)
df_lote_devolvido_PF['Ano de Óbito'] = df_lote_devolvido_PF['Ano de Óbito'].astype(int)

atributo = df_lote_devolvido_PF['Ano de Óbito']

condicao = (atributo.values < 2021) & (atributo.values>0)

cumpre = 2021 - atributo.values #O valor de 2021 foi escolhido para atribuir 1 se o óbito ocorreu em 2020

descumrpre = 0

df_lote_devolvido_PF['Tempo de Óbito'] = np.where(condicao, cumpre, descumrpre)
df_lote_devolvido_PF.groupby(['Ano de Óbito', 'Tempo de Óbito'])['Número'].count()
```

```python
df_lote_devolvido_PF = df_lote_devolvido_PF.drop(columns='Ano de Óbito')
```

```python
colunas_int_str = ['NI Beneficiário', 'Situação Cadastral - Código', 'País - Código', 'Estrangeiro', 
                   'Nacionalidade - Código', 'Sexo - Código', 'Município - Código', 'Telefone - DDD']
df_lote_devolvido_PF[colunas_int_str] = df_lote_devolvido_PF[colunas_int_str].astype(str)
df_lote_devolvido_PF.info()
```

### Quantidade de restituições por NI

```python
df_quant_ni = df_lote_devolvido_PF.groupby('NI Beneficiário')['Número'].count()
df_quant_ni
```

```python
percentuais = []
totais = []
devolvidas=[]
num_NIs=[]

for num in df_quant_ni.value_counts().index:
    filtro_nis = df_quant_ni[df_quant_ni==num].index
    num_NI = filtro_nis.shape[0]
    num_NIs.append(num_NI)
    total = df_lote_devolvido_PF[df_lote_devolvido_PF['NI Beneficiário'].isin(filtro_nis)]['Situação'].shape[0]
    totais.append(total)
    try:
        devolvida = df_lote_devolvido_PF[df_lote_devolvido_PF['NI Beneficiário'].isin(filtro_nis)]['Situação'].value_counts()['DEVOLVIDA']
        devolvidas.append(devolvida)
        percentuais.append(devolvida*100/total)
    except:
        percentuais.append(0)
        devolvidas.append(0)
```

### Devoluções por quartis

```python
percentuais = []
quantidades = []
for v in range(3,len(desc_valores)-1):
    if v < len(desc_valores)-2:
        inferior = desc_valores[v]
        superior = desc_valores[v+1]
        filtro1 = df_lote_devolvido_PF['Valor']>=inferior
        filtro2 = df_lote_devolvido_PF['Valor']<superior
        devolvidas = df_lote_devolvido_PF.loc[(filtro1)&(filtro2)]['Situação'].value_counts()['DEVOLVIDA']
        total = df_lote_devolvido_PF.loc[(filtro1)&(filtro2)]['Situação'].value_counts().sum()
        percentual = devolvidas*100/total
        percentuais.append(percentual)
        quantidades.append(total)
    else:
        inferior = desc_valores[v]
        superior = desc_valores[v+1]
        filtro1 = df_lote_devolvido_PF['Valor']>=inferior
        filtro2 = df_lote_devolvido_PF['Valor']<=superior
        devolvidas = df_lote_devolvido_PF.loc[(filtro1)&(filtro2)]['Situação'].value_counts()['DEVOLVIDA']
        total = df_lote_devolvido_PF.loc[(filtro1)&(filtro2)]['Situação'].value_counts().sum()
        percentual = devolvidas*100/total
        percentuais.append(percentual)
        quantidades.append(total)
        
percentis = [desc_valores.index[3] + ' - ' + desc_valores.index[4],
            desc_valores.index[4]+ ' - ' + desc_valores.index[5],
            desc_valores.index[5]+ ' - ' + desc_valores.index[6],
            desc_valores.index[6]+ ' - ' + desc_valores.index[7]]

pd.DataFrame(data={'Quartis':percentis, 'Quantidades':quantidades,  '% Devoluções':percentuais })
```
